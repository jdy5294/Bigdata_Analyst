### 1. 넘파이 배열 다루기

#### 1-2. ndarray 객체

```python
import numpy as np

# 배열 만들기
print(np.array([[1,2,3], [4,5,6]]))
```

```python
# 난수 이용 배열 생성
print(np.random.randn(2,3))
```

```python
# 원소 0 또는 1
print(np.zeros(5))
print(np.ones((2,3)))
```

```python
# arange와 reshape
ar1 = np.arange(20, 200, 10)     # 20부터 (200-1)까지 10씩 증가
ar2 = ar1.reshape(3, 6)           # 3*6배열로 재배치
print(ar1)
print(ar2)
```

```python
# 배열의 통계 메소드 사용
ar3 = np.array([[5,7,9], [-7,-6,19], [6,9,11]])
print(ar3, "\n")

print(ar3.sum())
print(ar3.mean())
print(ar3.max(), "\n")

print(ar3.max(axis = 0))          # 각 열의 원소 중 가장 큰 값을 구함
print(ar3.max(axis = 1))          # 각 행의 원소 중 가장 큰 값을 구함
```

```python
# 조건식을 사용한 연산
ar3 = np.array([[5,7,9], [-7,-6,19], [6,9,11]])
print(ar3 > 0, "\n")

# ar3의 원소가 0보다 작으면 -100으로 변경, 그렇지 않으면 ar3의 값으로 대입
ar4 = np.where(ar3 < 0, -100, ar3)
print(ar4)
```

```python
# 배열의 정렬
ar5 = np.array([[5,7,9], [-3,-6,19], [6,4,11]])
print(ar5, "\n")

ar5.sort(0)       # sort(0)는 열 단위로 원소를 정렬
print(ar5, "\n")

ar5.sort(1)       # sort(1)는 행 단위로 원소를 정렬
print(ar5)
```

### 3. 데이터 전처리

#### 3-1. 데이터 변환

```python
# 파생변수 생성
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Bigdata_Analysis/python_csv/df_sample.csv")

df['합계'] = df['중간'] + df['기말']

x = ['1반', '1반', '1반', '2반', '2반', '2반', '3반', '3반', '3반', '3반']
df['cls'] = x
# 위와 같은 코드 3가지 : 
# df.insert(6, "cls", x, True) # 6번 열에 리스트 x로 df에 cls 파생변수 생성
# df.assign(cls = x)
# df.loc[:, 'cls'] = x

df
```

- 파이썬에서 Z-score 구하는 방법들
    1. **numpy**: z = (x - mean()) / std()
    2. **scipy.stats**: zscore()
    3. **sklearn.preprocessing**: StandardScaler().fit_transform()

```python
# 표준화
mid_avg = np.mean(df['중간'])
mid_std = np.std(df['중간'])

df['중간_Z점수'] = (df['중간'] - mid_avg) / mid_std
df
```

- 정규화 (데이터의 범위를 0과 1사이로 변환하여 데이터의 분포 조정
    1. (x - 최소값) / (최대값 - 최소값)
    2. MinMaxScaler()

```python
fin_min = np.min(df['기말'])
fin_max = np.max(df['기말'])

df['기말_정규화'] = (df['기말'] - fin_min) / (fin_max - fin_min)
df
```

#### 3-4. 이상값 처리
**사분위범위**: 3사분위수 - 1사분위수

**이상치 판별**: (1사분위수 - 1.5 * 사분위범위) 보다 작거나 (3사분위수 + 1.5 * 사분위범위) 보다 큰 수

```python
df_1QR = df["기말"].quantile(0.25)
df_3QR = df["기말"].quantile(0.75)
df_IQR = df_3QR - df_1QR
print(df_1QR)
print(df_3QR)
print(df_IQR)
```
