```python
from itertools import combinations
from scipy.stats import pearsonr

num_feat = df_New.columns.values
comb_num_feat = np.array(list(combinations(num_feat, 2)))
corr_num_feat = np.array([])

for comb in comb_num_feat:
  #corr = np.corrcoef(df_New[comb[0]], df_New[comb[1]])[0][1]
  corr = pearsonr(df_New[comb[0]], df_New[comb[1]])[0]
  corr_num_feat = np.append(corr_num_feat, corr)

high_corr_num = comb_num_feat[corr_num_feat >= 0.9]
high_corr_num
```

```python
df_New.drop(high_corr_num[0][0], axis=1, inplace=True)
df_New
```

### 2. 제2유형

데이터셋을 이용하여 파킨슨병을 예측하는 모델을 로지스틱 회귀모형을 적용하여 생성하고, 이때 파킨슨병을 예측하는데 영향을 미치는 변수를 중요한 순서대로 3개 선정하시오.

이 모델에서 파킨슨병으로 진단하는 기준(thershold)을 0.5로 했을 때와 0.8로 했을 때의 F1-스코어를 비교하고 해석하시오.

- 분석조건
      - 'name' 컬럼 제거
      - 로지스틱 회귀를 위해 상수항 추가
      - 트레이닝셋과 테스트셋의 비율은 9:1
      - 모델의 최적화 방법론으로 "bfgs" 사용
      - 데이터 정규화는 min-max 스케일러 사용
      - Status는 카테고리 타입으로 변환

#### 2-1. 데이터 불러오기

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.linear_model import LogisticRegression

df = pd.read_csv("/content/drive/MyDrive/dataset/python_csv/Parkinsons.csv")
df.info()
```

#### 2-2. 데이터 전처리

```python
# name 변수 제거
df_preprocessing = df.drop('name', axis = 1, inplace=False)

# Normalization
scaler = MinMaxScaler()
df_processed = pd.DataFrame(scaler.fit_transform(df_preprocessing))

# 상수항 추가
df_processed = sm.add_constant(df_processed, has_constant='add')

# 열 이름 복구
df_processed.columns = np.append('const', df_preprocessing.columns.values)

df_processed
```

#### 2-3. 데이터 분석 수행

```python
X = df_processed.drop('status', axis=1, inplace=False)
y = df_processed['status'].astype('category')

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.1, random_state=2022)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model = sm.Logit(y_train, X_train)
results = model.fit(method='bfgs', maxiter=1000)
results.summary()
```

      p-value가 제일 작은 3개 {'spread2':0.093, 'RPDE':0.161, 'PPE':0.185}

#### 2-4. 임계치에 따른 모델 정확도 비교

```python
pred = results.predict(X_test)
pred80 = (pred>0.8).astype('int')
pred50 = (pred>0.5).astype('int')

print('임계치 0.8의 F1-스코어', f1_score(y_test, pred80))
print('임계치 0.5의 F1-스코어', f1_score(y_test, pred50))
```
